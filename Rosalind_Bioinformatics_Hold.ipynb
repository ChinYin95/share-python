#### Rabbits and Recurrence Relations

a and b represent 1 pair of rabbit respectively, range start from 2 because in the second month the rabbit have been grow up and start reproductive. Each pair produces 2 rabbit pairs. c represent the current total number of rabbits pair in the month. b means  parent rabbits pair. (k*b) means how many child rabbit pairs reproductive in that month. a represent the previous total number of rabbits pair.

```python
def fibonacci(n,k):
    a, b = 1, 1
    for i in range(2, n):
        c = a + k * b
        b = a
        a = c
    return c
# call back fibonacci function
print (fibonacci(32,2))
```

#### Mendel's First Law

```python
k1 = 21
m1 = 16
n1 = 27
# function of probability
def proba(k,m,n):
	pt = k+m+n                    # total of organism of AA, Aa, aa
	pt1 = pt -1.0                 # total of organism have used by one and least the other  
	p1 = k/pt*(k-1.0)/pt1*1.0     # first parent of the AA organism and second parent of AA organism 
	p2 = k/pt*m/pt1*1.0           # first parent of the AA organism and second parent of Aa organism
	p3 = k/pt*n/pt1*1.0           # first parent of the AA organism and second parent of aa organism
	p4 = m/pt*k/pt1*1.0           # first parent of the Aa organism and second parent of AA organism
	p5 = m/pt*(m-1.0)/pt1*0.75    # first parent of the Aa organism and second parent of Aa organism, 0.75 is producing dominant allele
	p6 = m/pt*n/pt1*0.5           # first parent of the Aa organism and second parent of aa organism, 0.5 is producing dominant allele
	p7 = n/pt*k/pt1*1.0           # first parent of the aa organism and second parent of AA organism
	p8 = n/pt*m/pt1*0.5           # first parent of the aa organism and second parent of Aa organism, 0.5 is producing dominant allele
	p9 = n/pt*(n-1.0)/pt1*0       # first parent of the aa organism and second parent of aa organism, there is no dominant allele
	probtot = p1+p2+p3+p4+p5+p6+p7+p8+p9  # all the probability of dominant allele have been added
	return probtot

# call back the function probability
print ("{:0.5f}".format(proba(k1,m1,n1)))
```

#### Consensus and Profile

```python
import sys
# function of dnaToProfile
def dnaToProfile( dna ):
    prof = [[0 for i in range(len(dna[0]))] for i in range(4)]
    # to read every column of the sequence alphabet
    for seq in dna:
        for i in range(len(seq)):
            if seq[i] == 'A':
                prof[0][i] += 1
            elif seq[i] == 'C':
                prof[1][i] += 1
            elif seq[i] == 'G':
                prof[2][i] += 1
            elif seq[i] == 'T':
                prof[3][i] += 1
    return prof

# function of profileToConsensus
def profileToConsensus( profile ):
    out = ''
    for i in range(len(profile[0])):
        # To find the max of every column alphabet
        m = max([profile[j][i] for j in range(4)])
        if profile[0][i] == m:
            out += 'A'
        elif profile[1][i] == m:
            out += 'C'
        elif profile[2][i] == m:
            out += 'G'
        elif profile[3][i] == m:
            out += 'T'
    return out
#To read the line from rosalind file
text = [line.strip() for line in open('rosalind_cons.txt')]
i = 2
while i < len(text):
    if not text[i].startswith('>') and not text[i-1].startswith('>'):
        text[i-1] += text[i]
        del text[i]
        i -= 1
    i += 1
# call the function dnaToProfile
profile = dnaToProfile([text[i] for i in range(1,len(text),2)])
# call the function of profileToConsensus and print
print (profileToConsensus(profile))
print ('A: ' + str(profile[0])[1:].replace(']','').replace(',',''))
print ('C: ' + str(profile[1])[1:].replace(']','').replace(',',''))
print ('G: ' + str(profile[2])[1:].replace(']','').replace(',',''))
print ('T: ' + str(profile[3])[1:].replace(']','').replace(',',''))
```

#### Mortal Fibonacci Rabbits

```python
# function of fibonacciRabbits
def fibonacciRabbits(n,m):
	F = [0, 1, 1]
	for i in range(3, n+1):
		if i<=m:
			total = F[i-1]+F[i-2]
		elif i == m+1:
			total = F[i-1]+F[i-2]-1
		else:
			total = F[i-1]+F[i-2]-F[i-m-1]
		F.append(total)
	return (F[n])

# call back fibonacciRabbits
print (fibonacciRabbits(98,16))
```

#### Inferring mRNA from Protein

```python
from functools import reduce
from operator import mul

freq = {
    'A': 4, 'C': 2, 'D': 2, 'E': 2,
    'F': 2, 'G': 4, 'H': 2, 'I': 3,
    'K': 2, 'L': 6, 'M': 1, 'N': 2,
    'P': 4, 'Q': 2, 'R': 6, 'S': 6,
    'T': 4, 'V': 4, 'W': 1, 'Y': 2,
    'STOP': 3
}
#function of load_dna_file
def load_dna_file(fname):
    with open(fname) as inf:
        dna = inf.read()
    return "".join(dna.split())   # removes all whitespace
# function of num_rna_strings
def num_rna_strings(dna, modulo=None):
    if modulo:
        reduce_fn = lambda a, b: (a * b) % modulo
    else:
        reduce_fn = mul
    freqs = (freq[base] for base in dna)
    return reduce(reduce_fn, freqs, freq["STOP"])

def main():
    dna = load_dna_file("rosalind_mrna.txt")
    num = num_rna_strings(dna, 1000000)
    print(format(num))

if __name__ == "__main__":
    main()
  
```

```python
DNA_CODON_TABLE = {
    'TTT': 'F',     'CTT': 'L',     'ATT': 'I',     'GTT': 'V',
    'TTC': 'F',     'CTC': 'L',     'ATC': 'I',     'GTC': 'V',
    'TTA': 'L',     'CTA': 'L',     'ATA': 'I',     'GTA': 'V',
    'TTG': 'L',     'CTG': 'L',     'ATG': 'M',     'GTG': 'V',
    'TCT': 'S',     'CCT': 'P',     'ACT': 'T',     'GCT': 'A',
    'TCC': 'S',     'CCC': 'P',     'ACC': 'T',     'GCC': 'A',
    'TCA': 'S',     'CCA': 'P',     'ACA': 'T',     'GCA': 'A',
    'TCG': 'S',     'CCG': 'P',     'ACG': 'T',     'GCG': 'A',
    'TAT': 'Y',     'CAT': 'H',     'AAT': 'N',     'GAT': 'D',
    'TAC': 'Y',     'CAC': 'H',     'AAC': 'N',     'GAC': 'D',
    'TAA': 'Stop',  'CAA': 'Q',     'AAA': 'K',     'GAA': 'E',
    'TAG': 'Stop',  'CAG': 'Q',     'AAG': 'K',     'GAG': 'E',
    'TGT': 'C',     'CGT': 'R',     'AGT': 'S',     'GGT': 'G',
    'TGC': 'C',     'CGC': 'R',     'AGC': 'S',     'GGC': 'G',
    'TGA': 'Stop',  'CGA': 'R',     'AGA': 'R',     'GGA': 'G',
    'TGG': 'W',     'CGG': 'R',     'AGG': 'R',     'GGG': 'G'
}

# function of translate_codon , translate three codon to one protein
def translate_codon(codon):
    protein = None
    if len(codon) == 3 and codon in DNA_CODON_TABLE:
        protein = DNA_CODON_TABLE[codon]
    return protein

# function of reverse_complement
def reverse_complement(dna):
    lookup = {'A':'T', 'T':'A', 'G':'C', 'C':'G'}
    return ''.join([lookup[c] for c in reversed(dna)])

# function of possible_protein_string, append one protein string
def possible_protein_strings(s):
    results = []
    indices = []

    l = len(s)
    for i in range(l):
        protein = translate_codon(s[i:i+3])
        if protein and protein == 'M':
            indices.append(i)

    for i in indices:
        found_stop = False
        protein_string = ''

        for j in range(i, l, 3):
            protein = translate_codon(s[j:j+3])

            if not protein:
                break

            if protein == 'Stop':
                found_stop = True
                break

            protein_string += protein

        if found_stop:
            results.append(protein_string)

    return results


if __name__ == "__main__":

    large_dataset ='GAGAGTGGAAGACCTAAACCCTTAACCGGGACTACCAACCTCGTTCTTTCGGTGTCCGTTTAGTATCGTGTTATGTGGTGTCAAGTACATAAAAGGTTGCCAACGCGGTGTGCCTCCGTTGGCGAATAGTAGGTAAAAGTGCATCACTGAGCCGTAGTGGGATATGCTACACGCCGACTGCTGACGTAGATTTGACGACCCCTTACAATCAATGATACCTGCATCCCAGCACTTTTCGGGTGAGTTTAAGACCGCTTACAAAATTAACTTAAAATTAGGAATCGCGCTAAGCTTTAAATTTTGTGAGTTGCTGGGATTGAATAGAGATATCGGTGTTGTCTGCTTGATCTAGGGATCTTTGCTGCGTGTCCTGCATGCAGAGCACGTTTACTGTACTTCTGATGGGGCGTATGGACAACGGATCACTCAGTCACATTTGGATTATCTCGCAACTCATGTCCGGCACCAACTTATGCCATTCGTAGCTACGAATGGCATAAGTTGGTGCCGGACATTTAGGGTCCGAGTCATGTTCCCAGGTGACTGCCAGTCGCTAGTATACTCCGACGGTGAGACTTCAGAGTTGGCACCTGCGTACGGAGGAGCGACAATCAATCCAACAGACCTTAAGATATCGATAGCGATCACGTTATGATCCCTAGTGGCAGCCGAGTACAACCTTCTGGACGCTATATTAATGCGTACTACATCGTGTACGTATGGGCCACGTATCTATCACGCTCTCTAGCTTAATTAGGCGCTAATTCCTCCCGCCCGGTAGATACTGGCGCTGTGGTGGCATACGGGTAGCGAAACTACATGATTACCCAGCCACTCTTAAAACTCACCCCCGACTGTTACCACATCCATCGAGAGGAGGGGGGCCAGATTACGTTGCATAAAAGAATCAAAACGTAAGAAGGGTCGCAACCCGGGGGGTTGTTTGCACGTGCACTACGGCTGGGTTGTC'
    # call back function of the possible_protein_string
    possible_a = possible_protein_strings(large_dataset)
    possible_b = possible_protein_strings(reverse_complement(large_dataset)) # call back function of reverse_complement
    print ("\n".join(set(possible_a + possible_b)))
```
